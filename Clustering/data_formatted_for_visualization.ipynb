# Generate Needed Datafiles for Visualization

## Cluster Breakdown Table for visualizing data from clustering run

```python
def get_column_wise_enrichment_list(df_clustering_summary_enrichment_analysis, df_summary_counts):

    all_enrichment_for_columns = []

    for cluster_id in range(0, len(df_summary_counts["cluster_id"])):
        
        enriched_go_terms = list(filter(None, list(df_clustering_summary_enrichment_analysis[cluster_id])))
        all_enrichment_for_columns.append(enriched_go_terms)

    return all_enrichment_for_columns
```

```python
def get_cluster_vis_summary(df_all_merged_clustering_data, df_clustering_summary_enrichment_analysis):
    
    df_all_merged_kmeans_data

    df_summary_counts = pd.DataFrame(pd.value_counts(df_all_merged_clustering_data['cluster_id']))

    df_summary_counts = df_summary_counts.reset_index()

    df_summary_counts.columns = ["cluster_id", "num_genes"]

    df_summary_counts = pd.DataFrame.sort_values(df_summary_counts, by = "cluster_id")

    df_summary_counts = df_summary_counts.reset_index()

    df_summary_counts = df_summary_counts.drop('index', axis = 1)

    column_wise_enrichment = get_column_wise_enrichment_list(df_clustering_summary_enrichment_analysis, df_summary_counts)

    df_summary_counts["enriched_go_terms"] = column_wise_enrichment

    return df_summary_counts
```

```python
def populate_ind_cluster_viz_data(clustering, data_save_prefix):

    num_clusters = len(clustering['cluster_id'].unique())
    
    for current_cluster in range(num_clusters): 
        
        # Obtain Loci data for each cluster in the clustering dataset chosen
        df_current_cluster_data = clustering[clustering['cluster_id'] == current_cluster]
        
        # Ensure desired file directory is made
        file_save_location = "../data/" + str(data_save_prefix) 
        
        if not os.path.exists(file_save_location):
            os.mkdir(os.path.join("../data/", data_save_prefix))
            
        os.mkdir(os.path.join(file_save_location, str(current_cluster)))
            
        
        
        # Save generated csv files for plots    
        
        # Gene List
        df_current_cluster_data.to_csv(file_save_location + "/" + str(current_cluster) + "/" + str(current_cluster) + "_gene_table.csv", index = False)
        
        # Cluster Metadata 
        df_metadata = pd.DataFrame([["name", str(current_cluster)], ["num_genes", len(df_current_cluster_data.index)]])
        df_metadata.to_csv(file_save_location + "/" + str(current_cluster) + "/" + str(current_cluster) + "_meta.csv", index = False)

        # Parallel Coordinate Plot Data
        df_differential_expression_data = df_current_cluster_data.T.iloc[0:22].T
        df_differential_expression_data.to_csv(file_save_location + "/" + str(current_cluster) + "/" + str(current_cluster) + "_diff_expression.csv", index = False)

        
```
